```python
import numpy as np
import pandas as pd
```

```python
games_table = pd.read_csv('https://raw.githubusercontent.com/patrick-mcclure/logistic_regression_tutorial/master/games.csv')
```

```python
games_table
```

```python
games_table.columns
```

```python
for o in ["firstBlood","firstTower","firstInhibitor","firstBaron","firstDragon","firstRiftHerald"]:
    column = games_table[o].to_numpy()
    games_table['t1_'+o] = (column == 1).astype(np.int32)
    games_table['t2_'+o] = (column == 2).astype(np.int32)
```

```python
objectives = [
 't1_firstTower',
 't2_firstTower',
 't1_firstInhibitor',
 't2_firstInhibitor',
 't1_firstBaron',
 't2_firstBaron',
 't1_firstDragon',
 't2_firstDragon',
 't1_firstRiftHerald',
 't2_firstRiftHerald',
 't1_firstBlood',
 't2_firstBlood',
 't1_towerKills',
 't2_towerKills',
 't1_inhibitorKills',
 't2_inhibitorKills',
 't1_baronKills',
 't2_baronKills',
 't1_dragonKills',
 't2_dragonKills']
```

```python
games_table
```

```python
winner_table = games_table[["winner"]]
```

```python
winner_table
```

```python
winner_array = winner_table.to_numpy()
```

```python
winner_array
```

```python
winner_array_binary = winner_array - 1
```

```python
winner_array_binary
```

```python
objectives_table = games_table[objectives]
```

```python
len(objectives)
```

```python
objectives_table
```

```python
objectives_array = objectives_table.to_numpy()
```

```python
objectives_array.shape
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
X_train, X_test, y_train, y_test = train_test_split(objectives_array,winner_array_binary,test_size=0.999, random_state=5683)
```

```python
games_train_table = pd.DataFrame(X_train,columns=objectives)
```

```python
games_train_table.insert(loc=0,column='winner',value=y_train+1) 
```

```python
games_train_table.to_csv('games_train.csv',index=False)
```

```python
games_test_table = pd.DataFrame(X_test,columns=objectives)
```

```python
games_test_table.insert(loc=0,column='winner',value=y_test+1) 
```

```python
games_test_table.to_csv('games_test.csv',index=False)
```
