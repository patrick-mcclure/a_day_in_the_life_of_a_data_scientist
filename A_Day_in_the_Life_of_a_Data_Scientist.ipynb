```python
import numpy as np
import pandas as pd
```

```python
#Load training data
games_table_train = pd.read_csv('https://raw.githubusercontent.com/patrick-mcclure/a_day_in_the_life_of_a_data_scientist/master/games_train.csv')
games_table_train
```

```python
#List input features
objectives = list(games_table_train.columns[1:])
objectives
```

```python
#Get input table for the training data
objectives_table_train = games_table_train[objectives]
objectives_table_train
```

```python
#Get train input data array
X_train = objectives_table_train.to_numpy()
```

```python
#Get target table
winner_table_train = games_table_train[['winner']]
winner_table_train
```

```python
#Get training target array
y_train = np.squeeze(winner_table_train.to_numpy())
y_train
```

```python
from sklearn.svm import LinearSVC
```

```python
#Train a model on the training data
model = LinearSVC(max_iter=10000)
model.fit(X_train,y_train)

#Predict the training data using the model
preds_train = model.predict(X_train)
acc_train = (preds_train == y_train).mean()
print('Acc_train: '+str(acc_train))
```

```python
#Load test data
games_table_test = pd.read_csv('https://raw.githubusercontent.com/patrick-mcclure/a_day_in_the_life_of_a_data_scientist/master/games_test.csv')
games_table_test
```

```python
#Get test data arrays, as done with training data
winner_table_test = games_table_test[["winner"]]
winner_array_test = winner_table_test.to_numpy()
y_test = np.squeeze(winner_array_test)
objectives_table_test = games_table_test[objectives]
X_test = objectives_table_test.to_numpy()
```

```python
#Predict the test data using the model
preds_test = model.predict(X_test)
acc_test = (preds_test == y_test).mean()
print('Acc_test: '+str(acc_test))
```

```python
import matplotlib.pyplot as plt
```

```python
#Plot model accuracies
fig, ax = plt.subplots()
xs = np.array(list(range(2)))
ax.set_ylabel('Accuracy',fontsize=14)
ax.set_ylim([0.75,1.01])
ax.set_yticks([0.8, 0.85, 0.9, 0.95, 1.0])
ax.tick_params(axis='y', labelsize=12)
ax.set_title('SVM Accuracies',fontsize=18)
ax.set_xticks(xs)
ax.set_xticklabels(['Train', 'Test'],fontsize=14)
ax.bar(xs,[acc_train, acc_test])
```

```python
from sklearn.model_selection import train_test_split
```

```python
#Get a validation split of the training data
X_train_val, X_test_val, y_train_val, y_test_val = train_test_split(X_train,y_train,test_size=0.25, random_state=1)
```

```python
#Perform grid search for regularization strength (i.e. c) using the validation split
for c in [10.0, 1.0, 0.1, 0.01, 0.001,0.0001,0.00001]:
    print(c)
    #Train a model on the validation training data using a regularization strength of c
    model_val = LinearSVC(penalty='l2', C=c,max_iter=10000)
    model_val.fit(X_train_val,y_train_val)
    
    #Predict the validation training data using the model with a regularization strength of c
    preds_train_val = model_val.predict(X_train_val)
    acc_train_val = (preds_train_val == y_train_val).mean()
    print('Acc_train_val: '+str(acc_train_val))
    
    #Predict the validation test data using the model with a regularization strength of c
    preds_test = model_val.predict(X_test_val)
    acc_test_val = (preds_test == y_test_val).mean()
    print('Acc_test_val: '+str(acc_test_val))
```

```python
#Train a model on the training data using a regularization strength of 0.001 
model_l2 = LinearSVC(penalty='l2', C=0.001)
model_l2.fit(X_train,y_train)

#Predict the training data using the model with a regularization strength of 0.001
preds_train_l2 = model_l2.predict(X_train)
acc_train_l2 = (preds_train_l2 == y_train).mean()
print('Acc_train: '+str(acc_train_l2))

#Predict the test data using the model with a regularization strength of 0.001
preds_test_l2 = model_l2.predict(X_test)
acc_test_l2 = (preds_test_l2 == y_test).mean()
print('Acc_test: '+str(acc_test_l2))
```

```python
#Plot the accuracies of the different models
fig, ax = plt.subplots()
xs = np.array(list(range(4)))
ax.set_ylabel('Accuracy',fontsize=14)
ax.set_ylim([0.75,1.01])
ax.set_yticks([0.8, 0.85, 0.9, 0.95, 1.0])
ax.tick_params(axis='y', labelsize=12)
ax.set_title('SVM Accuracies',fontsize=18)
ax.set_xticks(xs)
ax.set_xticklabels(['Train', 'Test', 'Train w/ L2', 'Test w/ L2'],fontsize=14)
ax.bar(xs,[acc_train, acc_test, acc_train_l2, acc_test_l2])
```

```python
#Plot input feature weights for the model with a regularization strength of 0.001
fig, ax = plt.subplots()
xs = np.array(list(range(len(objectives))))
ax.set_ylabel('Weight Value',fontsize=14)
ax.tick_params(axis='y', labelsize=12)
ax.set_title('SVM Weights',fontsize=18)
ax.set_xticks(xs)
ax.set_xticklabels(objectives,rotation = 90,fontsize=12)
ax.bar(xs,np.squeeze(model_l2.coef_))
```
